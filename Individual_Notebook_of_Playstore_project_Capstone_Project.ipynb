```python
def Mergedinfo():
    marv = pd.DataFrame(index=merged_apps_reviews.columns)
    marv['data_type'] = merged_apps_reviews.dtypes
    marv["count of non null values"] = merged_apps_reviews.count()
    marv['NaN values'] = merged_apps_reviews.isnull().sum()
    marv['% NaN values'] = (marv['NaN values']/len(merged_apps_reviews)).round(4)*100 
    marv['unique_count'] = merged_apps_reviews.nunique()
    return marv 
Mergedinfo()
```
